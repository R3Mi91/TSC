# Creación de la Tabla reporte

LEFT JOIN de Tabla2 con Tabla1

Realizamos un LEFT JOIN entre Tabla1 y Tabla2 usando la columna Material, selecciona Tienda y Material de Tabla1 y Nombre Material y Marca de Tabla2, con el fin de guardar el resultado en un DataFrame df_union para mostrarlo.

```python
query = """
SELECT
    t1.'Tienda',
    t1.'Material',
    t2.'Nombre Material',
    t2.'Marca'
FROM Tabla1 t1
LEFT JOIN Tabla2 t2 ON t1.Material = t2.Material
"""
df_union = pd.read_sql_query(query, conn)
display(df_union)
```

LEFT JOIN de Tabla4 con Tabla2

Ahora realizaremos un LEFT JOIN encadenado une Tabla1 con Tabla2 por Material y luego con Tabla4 por Marca, seleccionando Tienda y Material de Tabla1, Nombre Material y Marca de Tabla2, y nombre marca de Tabla4, guardando el resultado en df_2union y mostrándolo.

```python
query = """
SELECT *
FROM Tabla1 t1
LEFT JOIN Tabla2 t2 ON t1.Material = t2.Material
LEFT JOIN Tabla4 t4 ON t2.Marca = t4.marca
"""
df_2union = pd.read_sql_query(query, conn)
display(df_2union)
```

Finalmente crearemos de la tabla reporte completa sobre la que se harán las consultas

# Consultas

Consulta 1. Filtrado de las ventas de las tiendas que fueron menores a 4000 en el mes de abril de 2025

Obtiene filas únicas de REPORTE_TDA_MARCAS unidas con Tabla3, mostrando la tienda y sus ventas de abril 2025 solo si son menores a 4000, y ordena el resultado de mayor a menor ventas en df_distinct_filtered_report.

```python
query = """
SELECT DISTINCT
    rtm.Tienda,
    t3."nombre tienda",
    t3."Abril 2025" AS Ventas_Abril_2025
FROM REPORTE_TDA_MARCAS rtm
LEFT JOIN Tabla3 t3 ON rtm.Tienda = t3.Tienda
WHERE t3."Abril 2025" < 4000
ORDER BY Ventas_Abril_2025 DESC
"""
df_distinct_filtered_report = pd.read_sql_query(query, conn)
display(df_distinct_filtered_report)
```

Consulta 2. Reporte de las materiales que produce la marca de nombre AZOO y id 287

Filtra filas únicas de REPORTE_TDA_MARCAS donde Marca es 287 y el nombre de la marca contiene "AZOO", seleccionando Marca, nombre marca, Material y Nombre Material, y guarda el resultado en df_azoo_reporte.

```python
query = """
SELECT DISTINCT
    Marca,
    "nombre marca",
    Material,
    "Nombre Material"
FROM REPORTE_TDA_MARCAS
WHERE Marca = 287
AND "nombre marca" LIKE '%AZOO%'
"""
df_azoo_reporte = pd.read_sql_query(query, conn)
display(df_azoo_reporte)
```

Consulta 3. Reporte de las tiendas que venden marca ZOO MED.

Obtiene filas únicas de REPORTE_TDA_MARCAS unidas con Tabla3 donde Marca es 255 y el nombre marca es "ZOO MED", mostrando información de tienda, material, marca, distrito y estado, y guarda el resultado en df_zoo_med_reporte.

```python
query = """
SELECT DISTINCT
    rtm.Tienda,
    rtm.Marca,
    rtm."nombre marca",
    rtm.Material,
    rtm."Nombre Material",
    t3."nombre tienda",
    t3.Distrito,
    t3.Estado
FROM REPORTE_TDA_MARCAS rtm
LEFT JOIN Tabla3 t3 ON rtm.Tienda = t3.Tienda
WHERE rtm.Marca = 255
AND rtm."nombre marca" = 'ZOO MED'
"""
df_zoo_med_reporte = pd.read_sql_query(query, conn)
display(df_zoo_med_reporte)
```

Consulta 4. Reporte de todos los materiales que tienen una A como inicial y se venden en las tiendas de ID mayor a 7127

Filtra de REPORTE_TDA_MARCAS las filas donde Nombre Material empieza con "A" y Tienda es mayor a 7127, seleccionando solo Tienda y Nombre Material, y guarda el resultado en df_filtered_materials.

```python
query = """
SELECT
    Tienda,
    "Nombre Material"
FROM REPORTE_TDA_MARCAS
WHERE "Nombre Material" LIKE 'A%'
AND Tienda > 7127
"""
df_filtered_materials = pd.read_sql_query(query, conn)
display(df_filtered_materials)
```

Desde una perspectiva ejecutiva, los insights obtenidos a través de estas consultas revelan oportunidades concretas de mejora y decisión:



*   Tiendas con ventas menores a $4000 en abril 2025 → Identificadas mediante una
consulta SQL con WHERE "Abril 2025" < 4000, esta información permite focalizar estrategias comerciales en sucursales de bajo rendimiento. En SAS, se replicó con PROC SQL, asegurando trazabilidad del análisis.

*   Materiales producidos por marcas específicas (AZOO y ZOO MED) → Extraídos con LIKE '%AZOO%' y WHERE Marca = 287, estas consultas permiten mapear el portafolio de productos por marca. En SAS, se reconstruyó la lógica con DATA y PROC SQL, validando la consistencia entre entornos.

*    Materiales que comienzan con la letra A y se venden en tiendas con ID > 7127 → Esta segmentación, realizada con LIKE 'A%' AND Tienda > 7127, permite identificar productos con potencial de posicionamiento en tiendas de mayor tráfico. El mismo patrón fue replicado en SAS para asegurar la equivalencia de resultados.

*   Distribución geográfica de productos por marca y tienda → Mediante la unión de tablas con LEFT JOIN entre REPORTE_TDA_MARCAS y Tabla3, se enriqueció la base con atributos como distrito y estado, permitiendo visualizar la cobertura territorial. Esta estructura fue replicada en SAS con DATA y MERGE.
